## 3. 数据可视化

### 3.1 数据可视化简介
#### 数据可视化：数据与结论间建立联系
- 以高效性、建设性方式的讨论结果，**理解运营管理**与**业务性能**之间的关系，提供直接有效的**决策依据**。
- 在庞大复杂的资源环境中，有目的的**挖掘数据背后**的特征信息，揭示**不同思考角度下的潜在客户规律**。
 
目的：**解释数据  传递信息  压缩文信息  突出观点**
 - 比较两组数据的大小（直观比较）
 - 校验量数据的差异（是否来自同一个分布）
 - 统计随机变量的分布（如分位数3*区间等统计量可视化）
 
原则：**高效的表达 = 清晰地思路 = 有价值的工作**
 - 尽量用低维度表达
 - 避免视觉噪声（标记、注释、颜色）切忌喧宾夺主
 - 确保各个元素清晰

### 3.2 数据可视化的模式、对象与流程
### 3.2.1 数据可视化的模式
数据可视化的表现形式（模式）有三种：**交互式、交互呈现式和呈现式**
 - 交互式：用于引导/发现，服务于个体
 - 呈现式：用于展示/讲述，服务于群体
 
| 分类       | 用户交互         | 图像生成        |目标受众        | 传媒媒介        |
|:-----------:| :-------------:|:-------------:| :-------------:|:-------------:|
| 交互式 | 客户完全掌控数据  | 实时 |少量个体|网络交互软件|
| 交互呈现式 | 客户部分掌控数据  | 实时 |少量群体|网络自助设备|
| 呈现式 | 客户不能掌控数据  | 预先 |大量群体|视频演示文稿|

### 3.2.2 数据可视化的对象
- 定类数据：表现为类别，但不区分顺序，是由定类尺度计du量形成的。
- 定序数据：表现为类别，但有顺序，是由定序尺度计量形成的。
- 定距数据：表现为数值，可进行加、减运算，是由定距尺度计量形成的。
- 定比数据：表现为数值，可进行加、减、乘、除运算，是由定比尺度计量形成的。
    
    （后面两类可以统称为定量数据。）

- 定量数据（低维度优选）
  - 位置 > 长度 > 面积 > 体积 > 密度 > 颜色 
- 定序数据
  - 位置 > 密度 > 颜色 > 连接 > 包含
  - 密度可以通过疏密程度来体现，颜色通过深浅避免视觉噪声，连接可以用镜头等从属关系来体现有序性。
- 定类数据（类之间确保元素清晰）
  - 位置 > 颜色 > 连接 > 包含 > 形状

### 3.3.2 图片与多子图
一些时候要从多个角度进行图片对比，Matplotlib为此提出**子图**的概念：在较大的图形中同时放上一组较小的坐标轴。

这些子图可能是*画中画(inset)、网格图(grid of plots)或其他布局形式*。

在介绍创建子图之前，先介绍几个Matplotlib的对象`fig、ax、plt和其对应各样的subplot`，它们作用都极其类似。

```python
### 案例介绍
'''fig.add_subplot'''
fig = plt.figure()         # figure是matplotlib中最基础的一个对象，是一个总的画布
ax = fig.add_subplot(2,2,1)# add_subplot在画布中添加一个axes(可以理解为子区域)，并返回这个子区域参数的前两个表示子区域的行列数，最后一个表示子区域的顺序
ax.plot([1,2,3])           # 画图，里面是坐标点
'''plt.subplot'''
fig = plt.figure()
ax = plt.subplot(2,2,1)   # 和fig.add_subplot作用相同，只是直接调用plt.subplot，会获取当前活跃的figure对象，然后添加子区域
ax.plot([1,2,3])
'''fig,ax plt.subplots'''
fig,ax = plt.subplots(2,2)# 调用subplots会直接添加多个子区域，返回的ax是一个包含所有子区域的矩阵
ax1 = ax[0,0]             # 通过下标获取需要的子区域
ax1.plot([1,2,3])
```

### 上述小结
1. figure是matplotlib中最基础的一个对象，是一个总的画布，一般`fig = plt.figure()`
2. `fig.add_subplot`和`plt.subplot`作用相同，不过前者是*在画布中添加一个axes(可以理解为子区域)*，后者直接获取当前活跃的figure对象并添加子区域，均返回这个子区域参数的前两个表示子区域的行列数，最后一个表示子区域的顺序。(注意fig.add和plt的区别)
3. `fig,ax = plt.subplots(2,2)`调用subplots会直接添加多个子区域，返回的ax是一个包含所有子区域的矩阵，所以想看对应的子图需要通过下标`ax[0,0]`,`ax[0,1],ax[1,0],ax[1,1]`和子图对应。（fig和ax只是常用名，取个axes一样）
4. Matplotlib中，几乎所有的画图操作都是在`axes`上完成的,`plt.plot()`的源码实际上是获取当前的axes上面作图。

#### 手动创建子图

#### 简易的网格子图
- `plt.subplots(nrows,ncols,sharex,sharey,subplot_kw,**fig_kw)`:创建一个新的图片，报了已生成子图对象的Numpy数组。
 - `nrows`:subplot行数
 - `ncols`:subplot列数
 - `sharex`:所有subplot使用相同的X轴刻度，调价xlim影响所有。
  - `sharey`:同上作用于Y轴。
  - `subplot_w`:创建各subplot的关键字字典，用于生成子图。
  - `**fig_kw`:创建figure其他关键字，例如`figsize()`。
- `plt.subplot(n,m,i)`:将要创建的行数、列数和索引值，索引值从1开始，左上角到右下角依次增大。
- `plt.subplots_adjust(left,bottom,right,top,hspace,wspace)`:整理子图之间的距离，默认`None`.

```python
###网格子图案例：subplot()
fig = plt.figure()
fig.subplots_adjust(hspace=0.5,wspace=0.5)
for i in range(1,7):
    ax = fig.add_subplot(2,3,i)
    ax.text(0.5,0.5,str((2,3,i)),fontsize=18,ha='center')  #文本插入，后续介绍
```

#### 一行代码创建网格
创建一个大型网格子图时，向隐藏部分坐标轴时，subplots可以创建多个子图，并返回一个包含子图的Numpy数组。

```python
###网格子图案例2：subplots()
fig,axes = plt.subplots(2,3,sharex=True,sharey=True)

for i in range(2):
    for j in range(3):
        axes[i,j].text(0.5,0.5, str((i,j)),fontsize=18,ha='center')
```

#### pd.GridSpec:实现更复杂的排列方式
不规则的多行多列子图，可以结合`plt.GridSpec`函数。
- `plt.GridSpec([\nrows,ncols,wspace,hspace)`:可以复杂排列子图。跟`plt.subplots()`应用基本一样。
- 子图索引的行列可以取阶段，可能存在均分后，图大小不一，例如`[0,1:]`是均分之后的第1行第2张图，宽度包含了均分第2块及后续所有。

```python
### 案例说明
grid = plt.GridSpec(4,4,wspace=0.4,hspace=0.3)
plt.subplot(grid[:-1,1:])
plt.subplot(grid[:-1,0])
plt.subplot(grid[-1,1:])
```

### 3.3.3 文字注释
在图标上绘制自己的注释，可能包含文本、箭头和其它图形，可以使用`text、arrow、annote`方法，仅推荐`text和annote`。

注：虽然`plt.arrow()`函数可以实现这个功能，但因为创建出的箭头是SVG向量图对象，会随着图形分辨率的变化而变化，所以不推荐。
- `ax.text(x,y,s,fontdict=None,withdash=False,**kwargs)`:文本注释，只能填写文本。
 - `x,y`:坐标位置（标量）
 - `s`:注释的内容
 - `fontdict`:重新设置内容的文本格式，包括颜色、大小等（字典）
 - `withdash`:替代注释内容s的对象，破折号。
 - `fontsize`:字体大小。
 - `**kwargs`：其他参数,例如`ha`水平分布情况，`va`垂直分布情况。
- `ax.annote(text,xy,xytext,arrowprops,horizontalalignment,verticalalignment,**kwargs)`：箭头标注，更容易观看。
 - `text`:箭头标注的内容
 - `xy`:箭头尖端的位置
 - `xytext`:文本位置
 - `arrowprops`:箭头的参数设置，一般以`arrowprops = dict()`形式表示，内容包括：
  - `width`:宽度，以点为单位
  - `frac`:箭头头部占得比例
  - `facecolor`:颜色
  - `headwidth`:底部的宽度，以点为单位
  - `shrink`:移动提示，离注释点和文本的距离
  - `headlength`底部的长度，同上
 - `horizontalalignment`:水平位置，left、bottom、top、right位置
 - `verticalalignment`:垂直位置，同上

```python
### 案例说明
fig,ax = plt.subplots()

x = np.linspace(0, 20, 1000)
ax.plot(x,np.cos(x))
ax.axis('equal')            #限制图在坐标框架内

ax.annotate('local max', xy=(6.28,1), xytext=(10,4),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.annotate('local minimum', xy=(5 * np.pi, -1), xytext=(2,-6),
            arrowprops=dict(arrowstyle ="->",connectionstyle="angle3,angleA=0,angleB=-90"))
```

#### 箱型图科普
- Q1、Q2、Q3的含义
 - 确定四分位数的位置。`Qi所在位置=i（n+1）/4`，其中i=1，2，3。n表示序列中包含的项数。
 - Q1是下四分位数，Q2是中位数，Q3是上四分位数。
 - 案例说明：求12，15，17，19，20，23，25，28，30，33，34，35，36，37的Q1、Q2、Q3
  - Q1的位置 = (14+1)/4 = 3.75  Q1 = 0.25× 第三项 + 0.75×第四项 = 0.25 × 17 + 0.75 × 19 =18.5
  - Q2的位置 = 2（14+1）/4=7.5  Q2 = 0.5×第七项+0.5×第八项=0.5×25+0.5×28=26.5
  - Q3的位置 = 3（14+1）/4=11.25 Q3 = 0.75×第十一项+0.25×第十二项=0.75×34+0.25×35=34.25


- 上边缘（Q3+1.5IQR）、下边缘（Q1-1.5IQR）（其中 IQR=Q3-Q1 是 interquartile range）

（虚线延伸到两边的黑色线）

- 上四分位数（Q3）、下四分位数（Q1）（箱子上下蓝色的边框）：排序后第25%和第75%处的样本数值

- 中位数（箱子中间的红色线）：排序后第50%处的样本数值，不受异常值影响！

**注**：均值是所有数据的平均值，受异常值影响！

- 异常值（上下边缘以外的点被定义为异常值，算是异常值的阈值）

    处理异常值时与3σ标准的异同：统计边界是否受异常值影响、容忍度的大小
    btw：σ⇒68.2%，2σ⇒95.4%，3σ⇒99.7%
    3σ标准以均值和方差作为度量，异常值对其的影响是非常大的。若按照分位数刻画的箱式图来考察的话，*只要异常值不影响排序，那么异常值的大小是怎样都是什么关系的*，但**对于3σ标准来说，不管异常值是否影响排序，其值的任何变化都会显著影响方差σ**。

    所以，如果能估计出数据存在特别特别离谱的异常值时，建议可以通过箱式图来考察，去度量数据的统计特性、集中程度、分布情况等；若数据中没有太可能有太多异常值的时候，就可以用3σ标准来考察。

    例子：
    箱子很短，说明数据很集中；箱子很长，说明数据很分散；虚线很长，说明异常值不仅大，还非常多。所以，如果想截取一个干净的数据，就需要把分位数的上下阈值定的足够的窄才行。
    
    同一数轴上，几批数据的箱线图并行排列，**几批数据的中位数、尾长、异常值、分布区间等形状信息便昭然若揭**。如上图，可直观得看出第三季度各分公司的销售额大体都在下降。

    但箱形图也有他的*局限性*，比如：**不能精确地衡量数据分布的偏态和尾重程度**；对于批量比较大的数据，反映的信息更加模糊以及用中位数代表总体评价水平有一定的局限性。

#### 任务五：不同舱位等级的人生存分布情况

```python
### 不同舱位等级的人生存和死亡分布情况-堆积柱状图
def stackedbarplot(gb, y_data_names, colors, x_label, y_label, title):    
    # gb = 'Pclass'
    #将数据分为不同舱位的活着与死亡类别，根据已有的数量计算总的，和其分别的占比
    num_gb_surv_ornot = data[['Survived',gb]].groupby(gb).sum()    
    num_gb_surv_ornot['unservived'] = data[gb].value_counts() 
    num_gb_surv_ornot['Survived']    
    num_gb_surv_ornot['total'] = data[gb].value_counts()    
    num_gb_surv_ornot['survived_prop'] = num_gb_surv_ornot['Survived']/num_gb_surv_ornot['total']    
    num_gb_surv_ornot['unsurvived_prop'] = num_gb_surv_ornot['unservived']/num_gb_surv_ornot['total']    
    print(num_gb_surv_ornot)        
    x_data = [i+1 for i in range(len(data[gb].unique()))]   
    y_data_list = [num_gb_surv_ornot['survived_prop'],                   
                   num_gb_surv_ornot['unsurvived_prop']]        
    _, ax = plt.subplots()    
    #迭代绘制堆积柱状图
    for i in range(0, len(y_data_list)):        
        if i == 0:            
            ax.bar(x_data, y_data_list[i], color = colors[i],                    
                   align ='center', label = y_data_names[i])        
        else:
            # 采用堆积的方式，除了第一个分类，后面的分类都从前一个分类的柱状图接着画
            # 用归一化保证最终累积结果为1，下面bottom参数表示纵向从哪里开始画            
            ax.bar(x_data, y_data_list[i], color = colors[i],                    
                   bottom = y_data_list[i - 1], align = 'center',                    
                   label = y_data_names[i])    
        ax.set_xticks(np.linspace(1,len(num_gb_surv_ornot),
                                  len(num_gb_surv_ornot)))     
        ax.set_xticklabels(num_gb_surv_ornot.index)    
        ax.set_ylabel(y_label)    
        ax.set_xlabel(x_label)    
        ax.set_title(title)    
        ax.legend(loc = (1,1))             #在右上角外面
        
stackedbarplot(gb = 'Pclass'
               , y_data_names = ['Survived', 'Unservived']
               , colors = ['#539caf', '#7663b0']
               , x_label = 'Pclass'
               , y_label = 'Proportion of survived/unsurvived'
               , title = 'Proportion of survived/unsurvived by Pclass (1, 2, 3)')
```

```python
### 不同舱位等级的人年龄分布情况-箱型图
def boxplot(x_data, y_data, y_value, base_color, median_color):    
    x_feature = data[x_data].unique()    
    bp_data = []    
    for item in x_feature: 
        bp_data.append(y_data[y_data[x_data] == item] [y_value].values)    
    _, ax = plt.subplots()        
    ax.boxplot(bp_data,                
               patch_artist = True,               
               medianprops = {'color': base_color},               
               boxprops = {'color': base_color, 'facecolor': median_color},               
               whiskerprops = {'color': median_color},               
               capprops = {'color': base_color})        
    ax.set_xticklabels(x_feature)    
    ax.set_ylabel('Values of %s' % y_value)    
    ax.set_xlabel('Features of the %s '                   
                  % x_data)    
    ax.set_title('The box with %d features for the %s'                  
                 % (len(x_feature), y_value) )
    
boxplot(x_data = 'Pclass',
        y_data = data[['Pclass','Age']].dropna(),  #Age的缺失值较多
        y_value = 'Age',
        base_color = 'b',
        median_color = 'r')
```
